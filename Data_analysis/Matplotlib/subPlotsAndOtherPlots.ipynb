# Pie Chart 

## Used to represents parts of a whole relationship 

```python
salary = [200,800,400,600]
name = ['King', 'Cecil', 'Jakes','Queen']
```

```python
?plt.pie
```

```python
plt.figure(figsize=(7,7))
plt.pie(salary, labels=name, autopct="%.0f%%", shadow=True, explode=[0,0.2,0,0],textprops={'fontsize':9, 'color':'k'});
plt.legend();

plt.title("Display of the various salaries by workers")
```

# Bar chart

## Used for various comparism

```python
plt.bar(name,salary, color=['red','gold','green','k']);
plt.title("Comparing various salaries amongst workers");
plt.xlabel('Names')
plt.ylabel('Title')

plt.show()
```

## Horizontal bar display 

```python
plt.barh(name,salary, color='k');
plt.title("Comparing various salaries amongst workers");
plt.xlabel('Names')
plt.ylabel('Title')

plt.show()
```
