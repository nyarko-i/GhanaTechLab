```python
plt.figure(figsize=(10,5))
plt.plot([2,4,6,7,8,12,17],[4,8,12,14,16,24,34], color="g", marker='o', linestyle = '-')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid()
plt.title('First line plot ');
plt.savefig("Line_Plot")
#plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.plot(gas['Year'],gas['France'], color='red',marker="*", label='France')
plt.plot(gas['Year'],gas['UK'], color='k', marker='s' ,label="UK")
plt.plot(gas['Year'],gas['USA'], color='b', marker='+', label= "USA")
plt.xlabel("Years", fontsize=14)
plt.ylabel("Gas Prices",fontsize=14)
plt.grid()
plt.title('Gas Prices of various countries over the years', fontsize=18)
plt.xticks(gas['Year'], rotation=30);
plt.legend();
plt.savefig("Lineplot.png");
```
