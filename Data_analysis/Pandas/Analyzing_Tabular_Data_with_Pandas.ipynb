## Continuation of tabular data analysis 

# Groupby

```python
sales.groupby('Region').size()
```

```python
sales.groupby('Region')['Quantity'].sum()
```

```python
sales.groupby(by='Category')[['Quantity', 'Revenue', 'Unit Price']].mean()
```

```python
sales.groupby(by='Category')[['Quantity', 'Revenue', 'Unit Price']].sum().reset_index()
```

```python
sales.groupby(by='Category')[['Quantity', 'Revenue', 'Unit Price']].min().reset_index()
```

```python
sales.groupby(by='Category')[['Quantity', 'Revenue', ]].max().reset_index()
```

```python
sales.groupby(['Region', 'State'])[['Quantity']].sum().reset_index()
```
