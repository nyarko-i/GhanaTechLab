```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
from random import randint 
import seaborn as sns
```

## Plot activity 1. 
### Line plot

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas.head()
```

```python
plt.figure(figsize=(15,7))
plt.plot(gas['Year'],gas['Canada'], color='red',marker="*", label='Canada');
plt.plot(gas['Year'],gas['France'], color='yellow',marker="s", label='France')
plt.plot(gas['Year'],gas['Japan'], color='k',marker="+", label='Japan')
plt.plot(gas['Year'],gas['Australia'], color='green',marker="o", label='France')
plt.legend();
plt.grid()
plt.xticks(gas['Year'], rotation=30)
plt.xlabel('Years')
plt.ylabel("Countries")
plt.title('Gas prices of four countries over the years ')
plt.show()
```

# Pie plot - Activity 2. 

```python
workers=[40, 35, 100, 70]
Education=['Primary', 'JHS', 'SHS', 'Tertiary']
```

```python
plt.figure(figsize=(15,7))
plt.pie(workers, labels=Education, autopct=('%.0f%%'), shadow=True, explode=[0.2,0,0,0])
plt.legend()
plt.title("Various levels of Education")
plt.show()
```

 ## Scatter Plot - Activity 3 

```python
xt = np.array([0,0.1,0.2,0.3,0.5,0.8,1.0])
yt=np.array([1.0,2.1,3.5,6.5,7.2,5.9,6.3])
zt= xt*yt
```

```python

#1.Changing the shape of the points to square
sns.set_style('darkgrid')
plt.scatter(xt,yt, label=yt,marker='s')
plt.scatter(xt,zt, label=zt, marker='^')
plt.xlabel('xt')
plt.ylabel('Value')
plt.legend()
plt.title('Display of various values using scatter plot')
plt.show()
```

## Bar plot - Activity 4

```python
course=['matlab','R','Python','Excel','Java']
values=[20,15,57,60,30]
```

```python
plt.bar(course,values, color=['k','red','gold','green','m'])
plt.xlabel('Course')
plt.ylabel("Values")
plt.title("Display of Various courses and their Values")
plt.show()
```

## Hist plot - Activity 5

```python
marks=[randint(20,90) for i in range(50)]
bins=[40,50,60,70,80,90,100]
```

```python
plt.hist(marks,bins, color='k')
plt.xlabel('Bins')
plt.ylabel('Values')
plt.title('Histogram of Random values')
plt.show()
```

## Heat Map - Activity 6

```python
heat = gas.corr()
heat.head()
```

```python
sns.heatmap(data=heat, annot=True, cmap='twilight')
plt.suptitle("Heatmap of correlated gas prices ")
plt.show()
```

## Creating a subplot for all the above - Activity 7

```python
plt.figure(figsize=(20, 15))

plt.subplot(2,3,1)
plt.plot(gas['Year'],gas['Canada'], color='red',marker="*", label='Canada');
plt.plot(gas['Year'],gas['France'], color='yellow',marker="s", label='France')
plt.plot(gas['Year'],gas['Japan'], color='k',marker="+", label='Japan')
plt.plot(gas['Year'],gas['Australia'], color='green',marker="o", label='France')
plt.legend();
plt.grid()
plt.xticks(gas['Year'], rotation=30)
plt.xlabel('Years')
plt.ylabel("Countries")
plt.title('Gas prices of four countries over the years ')


plt.subplot(2,3,2)
plt.hist(marks,bins, color='k')
plt.xlabel('Bins')
plt.ylabel('Values')
plt.title('Histogram of Random values')


plt.subplot(2,3,3)
sns.set_style('darkgrid')
plt.scatter(xt,yt, label=yt,marker='s')
plt.scatter(xt,zt, label=zt, marker='^')
plt.xlabel('xt')
plt.ylabel('Value')
plt.legend()
plt.title('Display of various values using scatter plot')


plt.subplot(2,3,4)
plt.bar(course,values, color=['k','red','gold','green','m'])
plt.xlabel('Course')
plt.ylabel("Values")
plt.title("Display of Various courses and their Values")


plt.subplot(2,3,5)
plt.pie(workers, labels=Education, autopct=('%.0f%%'), shadow=True, explode=[0.2,0,0,0])
plt.legend()
plt.title("Various levels of Education")


plt.subplot(2,3,6)
sns.heatmap(data=heat, annot=True, cmap='twilight')
plt.suptitle("Heatmap of correlated gas prices ")


plt.tight_layout()
plt.suptitle("Various plot trends - Dashboard", fontsize=14);
plt.savefig('Dashboard')
```
